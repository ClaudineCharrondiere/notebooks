```python
# import modules
import xarray as xr
from matplotlib import pyplot as plt
import matplotlib.cm as cm
import matplotlib.colors as colors
from datetime import datetime
import numpy.ma as ma
import numpy as np
import netCDF4 as nc4
from scipy.io import netcdf
%matplotlib inline
```

```python
TRCfile = '/home/albert/Data/NATL60/NATL60-CJM165-S/1d/NATL60-CJM165.1d_curloverf.nc'
```

```python
chunks = (1727,2711)
xr_chunks = {'x': chunks[-1], 'y': chunks[-2]}
navlon = xr.open_dataset(TRCfile,chunks=xr_chunks,engine='netcdf4')['nav_lon']
navlat = xr.open_dataset(TRCfile,chunks=xr_chunks,engine='netcdf4')['nav_lat']
time = xr.open_dataset(TRCfile,chunks=xr_chunks,engine='netcdf4')['time_counter']
```

```python
t=1
TRC = xr.open_dataset(TRCfile,chunks=xr_chunks,engine='netcdf4')['socurloverf'][t,:,:]
fig=plt.figure(figsize=(20,15))
title='Surf curl over f '+str(t)+'e frame'
plt.pcolormesh(navlon,navlat,ma.masked_invalid(TRC),norm=colors.LogNorm(vmin=0.00001, vmax=100),cmap='jet')
plt.title(title)
plt.autoscale(tight=True)
plt.colorbar(orientation='horizontal')
plt.savefig('/home/albert/Data/NATL60/PLOTS/Curloverfsurf_'+str(t)+'.png')
plt.close(fig)
```

```python
for t in np.arange(0,265,1):
    plt.figure(figsize=(20,15))
    title='Surf TRC '+str(t)+'e frame'
    plt.pcolormesh(navlon,navlat,ma.masked_invalid(TRC[t,0,:,:]),norm=colors.LogNorm(vmin=0.00001, vmax=100),cmap='jet')
    plt.title(title)
    plt.autoscale(tight=True)
    plt.colorbar(orientation='horizontal')
    plt.savefig('/home/albert/Data/NATL60/PLOTS/TRCsurf_'+str(t)+'.png')
    
```
