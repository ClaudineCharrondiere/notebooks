```python
# import modules
import xarray as xr
from matplotlib import pyplot as plt
import matplotlib.cm as cm
import numpy.ma as ma
import numpy as np
import netCDF4 as nc4
from datetime import datetime
from scipy.io import netcdf
%matplotlib inline
```

```python
TRCfile = '/home/albert/Data/NATL60/NATL60-CJM165-S/1d/2013/NATL60-CJM165_y2013m10d01.1d_ptrcT_surf.nc'
```

```python
chunks = (1727,2711)
xr_chunks = {'x': chunks[-1], 'y': chunks[-2]}
TRC = xr.open_dataset(TRCfile,chunks=xr_chunks,engine='netcdf4')['TRC'][0,0]
navlon = xr.open_dataset(TRCfile,chunks=xr_chunks,engine='netcdf4')['nav_lon']
navlat = xr.open_dataset(TRCfile,chunks=xr_chunks,engine='netcdf4')['nav_lat']
```

```python
TRC.shape
navlon.shape
```

```python
plt.figure(figsize=(20,15))
plt.pcolormesh(navlon,navlat,ma.masked_invalid(TRC),vmin=0,vmax=1.,cmap='jet')
plt.autoscale(tight=True)
plt.colorbar(orientation='horizontal')
plt.savefig('/home/albert/Data/NATL60/PLOTS/TRCsurf_20131001.png')
```
