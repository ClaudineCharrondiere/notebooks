# Antartic Plot of Sea Surface Temperature

ESOUTH data are available on meolkerg : /fsnet/data/meom/MODEL_SET/ORCA12.L75/EORCA12.L75-MJMgd16-MEAN

```python
## Importation des modules et packages python nécessaires

import numpy as np
import xarray as xr
import os
from netCDF4 import Dataset
import matplotlib.cm as mplcm
import colormap as cmaps
import matplotlib.cm as cm
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
import cartopy.feature as cfeature

%matplotlib inline
```

```python
## Les données de modèle et le fichier de grille
sst_file="/media/extra/DATA/ESOUTH12/ESOUTH12.L75-MJMgd16_y1990-2015_1m_sst.nc"
berg_file="/media/extra/DATA/ESOUTH12/ESOUTH12.L75-MJMgd16_y1990-2015_1m_bergmelt.nc"
grid_file="/media/extra/DATA/ESOUTH12/ESOUTH12.L75-MJMgd16_mask.nc"

!ls $berg_file
!ls $sst_file
!ls $grid_file

# Ouverture des fichiers avec xarray
dsst=xr.open_dataset(sst_file)
dsmelt=xr.open_dataset(berg_file)
dsg=xr.open_dataset(grid_file)
```

```python
# Chargement de la grille en latitude et longitude
lon=dsg.nav_lon
lat=dsg.nav_lat

#Chargement de la Température de surface et du flux de fonte, 1er pas de temps du fichier (janvier 1990)
sst=dsst.votemper[0,0]
melt=dsmelt.berg_melt[0]
```

```python
##Le plot : une carte de la température de surface en projection cylindrique équidistante

#Taille de la figure
fig = plt.figure(figsize=(20,5))

#Choix de la projection
ax = fig.add_subplot(111,projection=ccrs.PlateCarree(central_longitude=-30))

#Choix de la colorbar
cmap = plt.cm.jet
cmap.set_under(color='grey')

#Le plot
pcolor=ax.pcolormesh(lon,lat,sst,transform=ccrs.PlateCarree(),cmap=cmap,vmin=-5,vmax=30)

#Le masque des continents
ax.add_feature(cfeature.LAND,facecolor='grey')
#La grille lat-lon
ax.coastlines()
ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True,
                  linewidth=2, color='gray', alpha=0.5, linestyle='--')

fig.subplots_adjust(right=0.8)

#Légendes latitude/longitude
ax.text(-0.07, 0.55, 'Latitude (in degree)', va='bottom', ha='center',
        rotation='vertical', rotation_mode='anchor',
        transform=ax.transAxes)
ax.text(0.5, -0.1, 'Longitude (in degree)', va='bottom', ha='center',
        rotation='horizontal', rotation_mode='anchor',
        transform=ax.transAxes)
#Position, orientation, taille et légende de la colorbar
cbar = plt.colorbar(pcolor,orientation='horizontal',shrink=0.75,label='deg C')

#Le titre
fig.suptitle("Surface Temperature January 1990", fontsize=25)
```

```python
##Le même plot en projection orthographique (=azimutale)

fig = plt.figure(figsize=(20,15))
ax = fig.add_subplot(111,projection=ccrs.Orthographic(central_longitude=-30,
                                                    central_latitude=-90))
cmap = plt.cm.jet
cmap.set_under(color='grey')
pcolor=ax.pcolormesh(lon,lat,sst,transform=ccrs.PlateCarree(),cmap=cmap,vmin=-5,vmax=30)
ax.set_global()
ax.add_feature(cfeature.LAND,facecolor='grey')
ax.coastlines()
cbar=plt.colorbar(pcolor,orientation='vertical',fraction=0.026,pad=0.1)
cbar.set_label("deg C",size=17)
cbar.ax.tick_params(labelsize=20)
fig.suptitle("Surface Temperature January 1990", fontsize=25)
```

```python
##Toujours en projection orthographique, un plot du flux d'eau douce dû à la fonte des icebergs

fig = plt.figure(figsize=(20,15))
ax = fig.add_subplot(111,projection=ccrs.Orthographic(central_longitude=-30,
                                                    central_latitude=-90))

#Une autre palette
cmap = plt.cm.Blues
cmap.set_under(color='white')

pcolor=ax.pcolormesh(lon,lat,melt,transform=ccrs.PlateCarree(),cmap=cmap,vmin=0,vmax=0.00002)
ax.set_global()
ax.add_feature(cfeature.LAND,facecolor='grey')
ax.coastlines()
cbar=plt.colorbar(pcolor,orientation='vertical',fraction=0.026,pad=0.1)
cbar.set_label("kg/m2/s",size=17)
cbar.ax.tick_params(labelsize=20)
fig.suptitle("Iceberg Melt January 1990", fontsize=25)
```
